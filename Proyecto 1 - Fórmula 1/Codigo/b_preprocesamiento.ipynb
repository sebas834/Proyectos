La regla del 107 % se introdujo para la temporada 1996 y estuvo en vigor hasta la temporada 2002 La regla del 107 % es una regulación deportiva que afecta a las sesiones de clasificación de las carreras de Fórmula 1

# Unión

```python
# circuits_cols: circuitId	circuitRef	location	country	lat	lng OK
# constructors_cols: constructorId	constructorRef	name	nationality OK
# drivers_cols: driverId	driverRef	forename	surname	birthdate	nationality OK
# lap_times_cols: raceId	driverId	lap	position	time 
# pit_stop_cols: raceId	driverId	stop	lap	duration
# races_cols: 	raceId	year	round	circuitId	name	date OK
# results_cols: 	resultId	raceId	driverId	constructorId	position	time	statusId OK 
# status: statusId	status OK
```

```python
# Unión de las tablas
df_f1 = pd.merge(results_cols.iloc[:,1:], races_cols, on = "raceId").merge(constructors_cols, on = "constructorId", suffixes=('_race','_constructor')).merge(circuits_cols, on = "circuitId").merge(drivers_cols, on = "driverId", suffixes=('_constructor','_driver')).merge(status, on = "statusId")
```

```python
# Nulos y duplicados
funciones.nudu(df_f1)
```

```python
df_f1_dropcols = df_f1.copy()
for cols in df_f1_dropcols.columns:
    if "Id" in cols:
        df_f1_dropcols.drop(cols, axis = 1, inplace=True)
```

La regla del 107 % se introdujo para la temporada 1996 y estuvo en vigor hasta la temporada 2002

```python
df_2002 = df_f1_dropcols[df_f1_dropcols.date_race.dt.year > 2002]
```
