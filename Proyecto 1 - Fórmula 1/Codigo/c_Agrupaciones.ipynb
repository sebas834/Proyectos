# Importar librerías

```python
import pandas as pd
```

# Lectura de datos

```python
df = pd.read_pickle('formula_1.pkl')
```

```python
df.dtypes
```

# Agrupaciones y visualizaciones

```python
df.head()
```

```python
# Los más más ganador por año a partir del 2000
df.query("position == '1'").groupby(['surname','year'])[['position']].count().sort_values("position", ascending = False)
```

```python
# Los más ganadores a partir del año 2000
df.query("position == '1'").groupby(['surname'])[["position"]].count().sort_values("position", ascending = False).head(7)
```

```python
# Cantidad de victorias de los conductores por carrera
df.query("position == '1'").groupby(['name_race','surname'])[['position']].count().sort_values('position', ascending=False).head(15)
```

```python
# Cantidad de campeonatos por corredor en United States Grand Prix en todos los años
df.query("position == '1' & name_race == 'United States Grand Prix'").groupby(['name_race','surname'])[['position']].count().sort_values('position', ascending=False)
```

```python
df.head()
```

```python
# Los contructores más ganadores a partir del año 2000
df.query("position == '1'").groupby(['name_constructor'])[["position"]].count().sort_values("position", ascending = False).head(10)
```
